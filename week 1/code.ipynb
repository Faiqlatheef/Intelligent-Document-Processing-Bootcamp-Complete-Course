```python
# Step 2: Import Libraries
import fitz  # PyMuPDF
import os
import cv2
import numpy as np
import pandas as pd
from paddleocr import PaddleOCR
from PIL import Image
from google.colab.patches import cv2_imshow
import json
```

```python
ROOT = "/content/drive/MyDrive/idp_bootcamp/week_1"
pdf_path = os.path.join(ROOT, "sri_lanka_electricity_32.pdf")
output_dir = os.path.join(ROOT, "pdf_pages")
os.makedirs(output_dir, exist_ok=True)
```

```python
# Step 4: Convert PDF Pages to Images
def convert_pdf_to_images(pdf_path, output_dir):
    doc = fitz.open(pdf_path)
    for page_num in range(doc.page_count):
        page = doc.load_page(page_num)
        pix = page.get_pixmap()
        image_path = os.path.join(output_dir, f"page_{page_num + 1}.png")
        pix.save(image_path)
        print(f"Saved {image_path}")

convert_pdf_to_images(pdf_path, output_dir)
```

```python
# Step 5: Initialize PaddleOCR
ocr = PaddleOCR(use_angle_cls=True, lang='en')  # Initialize PaddleOCR
```

```python
# Step 6: Define Text Extraction Function
def extract_text(image_path):
    ocr_result = ocr.ocr(image_path, cls=True)
    extracted_data = []
    for line in ocr_result:
        for word_info in line:
            box = word_info[0]
            text = word_info[1][0]
            confidence = word_info[1][1]
            extracted_data.append({
                "text": text,
                "confidence": confidence,
                "box": box
            })
    return extracted_data
```

```python
# Step 7: Define Structure Detection Function
def detect_structure(image_path, extracted_text):
    # Load the image in grayscale
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Apply adaptive thresholding
    thresh = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                   cv2.THRESH_BINARY_INV, 11, 2)

    # Detect contours
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Initialize structure flags
    contains_table = False
    contains_paragraph = False
    contains_graph = False

    # Heuristic for table detection based on contour size
    for contour in contours:
        x, y, w, h = cv2.boundingRect(contour)
        if w > 100 and h > 30:
            contains_table = True
            break

    # Heuristic for paragraph detection based on text density
    text_density = len(extracted_text)
    if text_density > 20:
        contains_paragraph = True

    # If neither table nor paragraph, assume graph
    if not contains_table and not contains_paragraph:
        contains_graph = True

    return {
        "contains_table": contains_table,
        "contains_paragraph": contains_paragraph,
        "contains_graph": contains_graph
    }
```

```python
# Step 8: Define Table Extraction Function
def extract_table_data(image_path):
    # Perform OCR with table detection
    ocr_result = ocr.ocr(image_path, cls=True)

    table_data = []
    for line in ocr_result:
        row_data = []
        for word_info in line:
            text = word_info[1][0]
            row_data.append(text)
        if row_data:
            table_data.append(row_data)

    return table_data
```

```python
# Step 9: Process Each Page and Save Output in JSON
def process_pdf_pages(output_dir, ROOT):
    json_output = []
    page_files = sorted([f for f in os.listdir(output_dir) if f.endswith('.png')],
                        key=lambda x: int(x.split('_')[1].split('.')[0]))

    for page_file in page_files:
        page_number = int(page_file.split('_')[1].split('.')[0])
        image_path = os.path.join(output_dir, page_file)

        # Extract text
        extracted_text = extract_text(image_path)

        # Detect structure
        structure_info = detect_structure(image_path, extracted_text)

        # Initialize page data
        page_data = {
            "page_number": page_number,
            "structure": structure_info,
            "extracted_text": extracted_text
        }

        # Extract table data if table is present
        if structure_info["contains_table"]:
            table_data = extract_table_data(image_path)
            page_data["table_data"] = table_data

        # Append to JSON output
        json_output.append(page_data)

        print(f"Processed Page {page_number}")

    # Save all data to a single JSON file
    json_output_path = os.path.join(ROOT, "extracted_data.json")
    with open(json_output_path, 'w') as json_file:
        json.dump(json_output, json_file, indent=4)

    print(f"All extracted data saved to {json_output_path}")

process_pdf_pages(output_dir, ROOT)
```
